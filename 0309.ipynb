# Answer Code

---



```python
def solution(board):    
    # (배열 board를 순회하면서) 수열 A(i,j)을 구한다.
    for i in range(1, len(board)):
        for j in range(1, len(board[0])):
            if board[i][j]:            	
                board[i][j] = min(board[i - 1][j], board[i][j - 1], board[i - 1][j - 1]) + 1
    
    # (배열 board를 순회하면서) 1로 이루어진 가장 큰 정사각형의 한 변의 길이(i.e., max(A(i,j)) max_A를 구한다.
    max_A = 0          
    for row in board:
        if max_A < max(row):
            max_A = max(row)

    # (max_A를 활용하여) 1로 이루어진 가장 큰 정사각형의 넓이를 반환한다.                
    return max_A ** 2
```
